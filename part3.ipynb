```python
import cv2
import numpy as np
from keras.models import load_model
```

```python
model = load_model("facedetector.h5")
face_clsfr = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
labels_dict ={0:'mask_weared_incorrect',1:'without_mask',2:'with_mask'}
color_dict = {0:(0,0,255),1:(255,0,0),2:(0,255,0)}
```

cv2.CascadeClassifier : This is a class in the opencv library used for object detection using haar_like features.its specifically designed



```python
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    if not ret:
        break
    
    # Convert frame to grayscale (Haar cascade works on grayscale images)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_clsfr.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        face_img = gray[y:y + h, x:x + w]
        resized = cv2.resize(face_img, (100, 100))
        normalized = resized / 255.0
        reshaped = np.reshape(normalized, (1, 100, 100, 1))
        result = model.predict(reshaped)
        
        # Get the label with the highest prediction score
        label = np.argmax(result, axis=1)[0]
        
        # Draw a rectangle around the face
        cv2.rectangle(img, (x, y), (x + w, y + h), color_dict[label], 2)
        
        # Put the label text
        cv2.putText(img, labels_dict[label], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color_dict[label], 2)
    
    # Display the image
    cv2.imshow('Mask Detector', img)
    
    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
